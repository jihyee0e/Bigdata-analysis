#### PROJECT#8. 특정 주소에 대한 지리 정보 분석 후, 지도에 시각화

```python
import pandas as pd
```

```python
cbdf = pd.read_csv('./CoffeeBean_24.csv', encoding='cp949', header=0, engine='python', sep=',')

cbdf
```

```python
#오류 방지를 위해 필요한 column만 추출해서 다시 저장하기
cbdf_new = cbdf[['store_name', 'address']]
#cbdf_new = cbdf.iloc[:, 1:3]
#cbdf_new = cbdf.drop(['Unnamed: 0', 'phone'], axis=1)
cbdf_new
```

```python
addr = []
for address in cbdf_new.address:
    addr.append(str(address).split())  #주소(address)를 문자열로 받아서 띄어쓰기로 나누어 저장

addr
```

```python
#행정구역 주소 체계에 맞게 정리
addr2 = []

for i in range(len(addr)):
    if (addr[i][0] == '서울') | (addr[i][0]== '서울시'):
        addr[i][0] = '서울특별시'
    elif addr[i][0] == '부산시':
        addr[i][0] = '부산광역시'
    elif addr[i][0] == '인천':
        addr[i][0] = '인천광역시'
    elif addr[i][0] == '광주':
        addr[i][0] = '광주광역시'
    elif addr[i][0] == '대전시':
        addr[i][0] = '대전광역시'
    elif addr[i][0] == '울산시':
        addr[i][0] = '울산광역시'
    elif addr[i][0] == '세종시':
        addr[i][0] = '세종특별자치시'
    elif addr[i][0] == '경기':
        addr[i][0] = '경기도'
    elif addr[i][0] == '충북':
        addr[i][0] == '충청북도'
    elif addr[i][0] == '충남':
        addr[i][0] = '충청남도'
    elif addr[i][0] == '전북':
        addr[i][0] = '전라북도'
    elif addr[i][0] == '전남':
        addr[i][0] = '전라남도'
    elif addr[i][0] == '경북':
        addr[i][0] = '경상북도'
    elif addr[i][0] == '경남':
        addr[i][0] = '경상남도'
    elif (addr[i][0] == '제주') | (addr[i][0] == '제주시') | (addr[i][0] == '제주도'):
        addr[i][0] = '제주특별자치도'
    addr2.append(' '.join(addr[i]))

#len(addr)
addr2
```

```python
type(addr2)
```

```python
addr2 = pd.DataFrame(addr2, columns=['newAddress'])  #list타입을 그대로 붙일 수 없어서 pandas로 바꿔주기
addr2.head()

#type(addr2)
```

```python
cbdf2 = pd.concat([cbdf_new, addr2], axis=1)  #기존 파일에 행정구역 주소체계에 맞게 정리된 새 주소 합치기
cbdf2.head()
```

```python
cbdf2.to_csv('./CoffeeBean2.csv', encoding='cp949', index=False)
```

```python
mapcb = folium.Map(location=[37.491145, 126.9737511], zoom_start=15)

for i , store in cbGeodf.iterrows():
    #위도, 경도
    folium.Marker(location=[store['_Y'], store['_X']], popup=store['store_name'], 
                  icon=folium.Icon(color='red', icon='star')).add_to(mapcb)

mapcb.save('./map_cb.html')
```
