```python
import pandas as pd
import folium
```

```python
df = pd.read_csv('./보건복지부_보건의료기관현황.csv', encoding='cp949', header=0, 
                 engine='python', sep=',')
df
```

```python
#필요한 컬럼만 추출해서 다시 저장
new_df = df[['병원 명', '주소']]
new_df
```

```python
new_df.rename(columns={'병원 명':'병원명'}, inplace=True)  #띄어쓰기 수정
new_df
```

```python
addr = []
for address in new_df.주소:
    addr.append(str(address).split())  #주소 띄어쓰기로 나누어 저장하기

addr
```

- 열 수정하기: '시도'가 아닌 데이터 찾아서 수정

```python
#sido.iloc[27] = ['경상남도', '창원시']
#sido.iloc[31] = ['경상남도', '창원시']

#sido.iloc[47] = ['경상북도', '경산시']

#sido.iloc[209] = ['충청남도', '천안시']
#sido.iloc[210] = ['충청남도', '천안시']

addr[27].insert(0, '경상남도')
addr[31].insert(0, '경상남도')

addr[47].insert(0, '경상북도')

addr[209].insert(0, '충청남도')
addr[210].insert(0, '충청남도')

addr
```

```python
#행정구역 주소 체계에 맞게 정리
addr2 = []

for i in range(len(addr)):
    if (addr[i][0] == '서울') | (addr[i][0]== '서울시'):
        addr[i][0] = '서울특별시'
    elif addr[i][0] == '경기':
        addr[i][0] = '경기도'
    elif addr[i][0] == '부산시':
        addr[i][0] = '부산광역시'
    elif addr[i][0] == '인천':
        addr[i][0] = '인천광역시'
    elif addr[i][0] == '광주':
        addr[i][0] = '광주광역시'
    elif addr[i][0] == '대전시':
        addr[i][0] = '대전광역시'
    elif addr[i][0] == '울산시':
        addr[i][0] = '울산광역시'
    elif addr[i][0] == '세종시':
        addr[i][0] = '세종특별자치시'
    elif addr[i][0] == '충남':
        addr[i][0] = '충청남도'
    elif addr[i][0] == '전북':
        addr[i][0] = '전라북도'
    elif addr[i][0] == '전남':
        addr[i][0] = '전라남도'
    elif addr[i][0] == '경북':
        addr[i][0] = '경상북도'
    elif addr[i][0] == '경남':
        addr[i][0] = '경상남도'
    elif addr[i][0] == '충북':
        addr[i][0] == '충청북도'
    elif (addr[i][0] == '제주') | (addr[i][0] == '제주시') | (addr[i][0] == '제주도'):
        addr[i][0] = '제주특별자치도'

    addr2.append(' '.join(addr[i]))

addr2
```

```python
type(addr2)
```

```python
addr2 = pd.DataFrame(addr2, columns=['newAddress'])  #list타입을 그대로 붙일 수 없어서 pandas로 바꿔주기
addr2.head()
```

```python
new_df2 = pd.concat([new_df, addr2], axis=1)  #기존 파일에 행정구역 주소체계에 맞게 정리된 새 주소 합치기
new_df2.head()
```

```python
#파일 저장
#new_df2.to_csv('E:/0401/빅데이터분석/week11/hospital_address.csv', encoding='cp949', index=False)
new_df2.to_csv('./hospital_address.csv', encoding='cp949', index=False)
```
